```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Example data - replace with pd.read_csv('your_file.csv') for actual use
data = {
    'Load': ['100M', '200M', '300M', '400M', '500M', '600M', '700M', '800M', '900M', '1G'],
    'Bare Metal': [77, 86, 93, 94, 96, 96, 97, 99, 100, 101],
    'VM': [93, 96, 98, 101, 103, 111, 147, 140, 143, 153],
    'Container': [78, 87, 94, 94, 95, 95, 95, 95, 106, 107]
}

df = pd.DataFrame(data)

# Plotting
bar_width = 0.25
x = np.arange(len(df['Load']))

plt.figure(figsize=(14, 8))  # Increased height
plt.bar(x - bar_width, df['Bare Metal'], width=bar_width, label='Bare Metal', color='#d4aa00')
plt.bar(x, df['VM'], width=bar_width, label='VM', color='#d95f02')
plt.bar(x + bar_width, df['Container'], width=bar_width, label='Container', color='#e7298a')

# Bigger labels and tick sizes
plt.xlabel('Throughput Load', fontsize=16)
plt.ylabel('Power Consumption (W)', fontsize=16)
plt.title('Power Consumption vs Throughput Load', fontsize=18)
plt.xticks(x, df['Load'], fontsize=14)
plt.yticks(fontsize=14)
plt.legend(fontsize=14)

plt.tight_layout()
plt.show()
```
